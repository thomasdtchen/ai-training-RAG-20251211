```python
tools = load_tools(["llm-math", "wikipedia"], llm=llm)
```

### 创建Agent实例

```python
agent = create_react_agent(llm, tools)
```

```python
agent.invoke({"messages": [{"role": "human", "content": "1234的32.5%是多少？"}]})
```

```python
for event in agent.stream({"messages": [{"role": "human", "content": "1234的32.5%是多少？"}]}):
    print(event)
```

```python
for event in agent.stream({"messages": [{"role": "human", "content": "西安的特色美食是哪些？"}]}):
    print(event)
```
